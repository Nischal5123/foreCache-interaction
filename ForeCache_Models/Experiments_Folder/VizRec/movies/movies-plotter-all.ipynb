```python
import pandas as pd

# Load the uploaded CSV file
data = pd.read_csv('/Users/aryal/Desktop/ForeCache/foreCache-interaction/ForeCache_Models/Experiments_Folder/VizRec/birdstrikes/all_experiments_BirdStrikes.csv')
data['Algorithm'].replace({'ActorCritic': 'Actor-Critic', 'QLearn': 'Q-Learning'}, inplace=True)



# Display the filtered DataFrame
data.head(50)
```

```python
import matplotlib.pyplot as plt

def create_plot(average_accuracies, svg_file_path, openended):

  # Define the strategies list
    strategies = ['Greedy', 'Q-Learning', 'Bayesian', 'Momentum', 'HMM', 'Random', 'WSLS', 'Reinforce', 'SARSA', 'Actor-Critic']
    markers = ['s', 'o', '+', 'H', '*', 'D', '^', 'x', 'p', '|']
    # Extract values from the DataFrame and store them in the values list
    # strategies = ['Greedy', 'Q-Learning', 'Bayesian', 'Momentum', 'HMM']
    # markers = ['s', 'D', '+', 'H', '*']
    values = average_accuracies.loc[strategies].values.tolist()
    # X-axis percentages and labels
    x_labels = ['10%', '20%', '30%', '40%', '50%', '60%', '70%', '80%', '90%']



    # Plotting the new data with Y-axis limit up to 0.70
    plt.figure(figsize=(10, 15))

    for i, strategy in enumerate(strategies):
        plt.plot(values[i], marker=markers[i % len(markers)], markersize=12,label=strategy, linewidth=6)  # Increase line thickness

    plt.xlabel('Percentage of data used for training', fontsize=25, fontweight='bold')  # Increase font size and make it bold
    plt.ylabel('Accuracy (%)', fontsize=25, fontweight='bold')  # Increase font size and make it bold
    plt.xticks(range(len(x_labels)), x_labels, fontsize=20, fontweight='bold')  # Increase font size and make it bold for x-axis labels
    plt.yticks(fontsize=20, fontweight='bold')  # Increase font size and make it bold for y-axis labels
    # if openended == 'Openended':
    #     plt.legend(loc='upper left', bbox_to_anchor=(1, 1), fancybox=True, shadow=True, ncol=1, fontsize=15)
    plt.legend(loc='upper left', bbox_to_anchor=(1, 1), fancybox=True, shadow=True, ncol=1, fontsize=25)  # Increase font size for legend
    plt.title('Birdstrikes - ' +str(openended), fontsize=30, fontweight='bold')  # Increase font size and make it bold
    plt.grid(axis='y', linewidth=2)  # Increase grid line thickness

    plt.savefig(svg_file_path, format='svg', bbox_inches='tight')

    return svg_file_path
```

```python
data_focused=data[data['Openended']=='Focused']
average_accuracies_focused = data_focused.groupby(['Algorithm', 'Threshold']).Accuracy.mean().unstack()
average_accuracies_focused.drop(columns=[0.05], inplace=True)
# Display the calculated averages
average_accuracies_focused
create_plot(average_accuracies_focused, 'birdstrikes-focused-all.svg', 'Focused')
```

```python
data_open=data[data['Openended']=='Openended']
average_accuracies_open = data_open.groupby(['Algorithm', 'Threshold']).Accuracy.mean().unstack()
average_accuracies_open.drop(columns=[0.05], inplace=True)
# Display the calculated averages
average_accuracies_open
create_plot(average_accuracies_open, 'birdstrikes-open-all.svg', 'Openended')
```

```python
import matplotlib.pyplot as plt

def create_plot_index(average_accuracies, svg_file_path, openended):

   # Define the strategies list
    strategies = ['Greedy', 'Q-Learning', 'Bayesian', 'Momentum', 'HMM', 'Random', 'WSLS', 'Reinforce', 'SARSA', 'Actor-Critic']
    markers = ['s', 'o', '+', 'H', '*', 'D', '^', 'x', 'p', '|']
    # Extract values from the DataFrame and store them in the values list
    # strategies = ['Greedy', 'Q-Learning', 'Bayesian', 'Momentum', 'HMM']
    # markers = ['s', 'o', '+', 'H', '*']
    values = average_accuracies.loc[strategies].values.tolist()

    # X-axis percentages and labels
    x_labels = ['10%', '20%', '30%', '40%', '50%', '60%', '70%', '80%', '90%']



    # Plotting the new data with Y-axis limit up to 0.70
    plt.figure(figsize=(20, 15))

    for i, strategy in enumerate(strategies):
        plt.plot(values[i], marker=markers[i % len(markers)], markersize=15,label=strategy, linewidth=6)  # Increase line thickness

    plt.xlabel('Percentage of data used for training', fontsize=30, fontweight='bold')  # Increase font size and make it bold
    plt.ylabel('Accuracy (%)', fontsize=30, fontweight='bold')  # Increase font size and make it bold
    plt.xticks(range(len(x_labels)), x_labels, fontsize=20, fontweight='bold')  # Increase font size and make it bold for x-axis labels
    plt.yticks(fontsize=20, fontweight='bold')  # Increase font size and make it bold for y-axis labels
    # if openended == 'Openended':
    #     plt.legend(loc='upper left', bbox_to_anchor=(1, 1), fancybox=True, shadow=True, ncol=1, fontsize=15)
    legend = plt.legend(loc='upper left', bbox_to_anchor=(-0.5, -0.2), fancybox=True, shadow=True, ncol=10, fontsize=20)  # Increase font size for legend
    frame = legend.get_frame()
    frame.set_linewidth(4)  # Increase border thickness around legend

    plt.title('Movies - ' +str(openended), fontsize=35, fontweight='bold')  # Increase font size and make it bold
    plt.grid(axis='y', linewidth=4)  # Increase grid line thickness

    plt.savefig(svg_file_path, format='svg', bbox_inches='tight')

    return svg_file_path
```

```python
data_focused=data_movies[data_movies['Openended']=='Openended']
average_accuracies_focused = data_focused.groupby(['Algorithm', 'Threshold']).Accuracy.mean().unstack()
average_accuracies_focused.drop(columns=[0.05], inplace=True)
# Display the calculated averages
average_accuracies_focused
create_plot_index(average_accuracies_focused, 'index-all.svg', 'Focused')
```
