```python
import pandas as pd

# Load the uploaded CSV file
file_path = 'all_experiments_BirdStrikes.csv'
data = pd.read_csv(file_path)
data['Algorithm'].replace({'ActorCritic': 'Actor-Critic', 'QLearn': 'Q-Learning'}, inplace=True)

# Display the first few rows of the dataframe to understand its structure
data.head(50)
```

```python
import matplotlib.pyplot as plt

def create_plot(average_accuracies, svg_file_path, openended):

    # Define the strategies list
    strategies = ['Random', 'Greedy', 'WSLS', 'Q-Learning', 'Reinforce', 'SARSA', 'Actor-Critic', 'Bayesian', 'Momentum', 'HMM']
    # Extract values from the DataFrame and store them in the values list
    values = average_accuracies.loc[strategies].values.tolist()


    # X-axis percentages and labels
    x_labels = ['10%', '20%', '30%', '40%', '50%', '60%', '70%', '80%', '90%']
    markers = ['o', 's', '^', 'D', 'x', 'p', '*', '+']

    # Plotting the new data with Y-axis limit up to 0.70
    plt.figure(figsize=(8, 6))

    for i, strategy in enumerate(strategies):
        plt.plot(values[i], marker=markers[i % len(markers)], label=strategy)

    plt.xlabel('Percentage of data used for training', fontsize=15)
    plt.ylabel('Accuracy (%)', fontsize=15)
    plt.xticks(range(len(x_labels)), x_labels)  # Set custom x-axis ticks and labels
    # plt.yticks([i * 0.1 for i in range(8)])  # Setting Y-axis ticks up to 0.70
    # plt.ylim(0.0, 0.80)  # Setting Y-axis limit up to 0.70
    # if openended == 'Openended':
    #     plt.legend(loc='upper left', bbox_to_anchor=(1, 1), fancybox=True, shadow=True, ncol=1, fontsize=15)
    plt.title('Birdstrikes - ' +str(openended), fontsize=15)
    plt.grid(axis='y')
    # plt.show()

    plt.savefig(svg_file_path, format='svg', bbox_inches='tight')

    svg_file_path
```

```python
data_focused=data[data['Openended']=='Focused']
average_accuracies_focused = data_focused.groupby(['Algorithm', 'Threshold']).Accuracy.mean().unstack()
average_accuracies_focused.drop(columns=[0.05], inplace=True)
# Display the calculated averages
average_accuracies_focused
create_plot(average_accuracies_focused, 'birdstrikes-focused-all.svg', 'Focused')
```

```python
data_open=data[data['Openended']=='Openended']
average_accuracies_open = data_open.groupby(['Algorithm', 'Threshold']).Accuracy.mean().unstack()
average_accuracies_open.drop(columns=[0.05], inplace=True)
# Display the calculated averages
average_accuracies_open
create_plot(average_accuracies_open, 'birdstrikes-open-all.svg', 'Openended')
```

```python
def create_legend_svg(strategies, svg_file_path):
    # Create a figure and axis for the legend
    fig, ax = plt.subplots(figsize=(8, 1))


    # Plotting the legend horizontally
    handles = [plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='blue', markersize=10) for _ in range(len(strategies))]
    ax.legend(handles, strategies, loc='lower left', mode='expand', ncol=len(strategies), fontsize=10)

    # Save the legend as SVG
    plt.savefig(svg_file_path, format='svg', bbox_inches='tight')

# Call the function to create the legend SVG
strategies = ['Random', 'Greedy', 'WSLS', 'Q-Learning', 'Reinforce', 'SARSA', 'Actor-Critic', 'Bayesian', 'Momentum', 'HMM']
svg_file_path = 'legend.svg'
create_legend_svg(strategies, svg_file_path)
```
