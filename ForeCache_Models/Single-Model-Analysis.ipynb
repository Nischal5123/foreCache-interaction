```python
import seaborn as sns
import matplotlib.pyplot as plt

# Function to add accuracy values on the bars
def add_values_on_bars(ax):
    for p in ax.patches:
        ax.annotate(f'{p.get_height():.2f}', (p.get_x() + p.get_width() / 2., p.get_height()),
                    ha='center', va='center', fontsize=9, color='black', xytext=(0, 5),
                    textcoords='offset points')

# Plot accuracy by model and task with values on the bars
plt.figure(figsize=(10, 6))
ax = sns.barplot(data=df, x='Task', y='Accuracy', hue='Model')
plt.title('Accuracy by Model and Task')
add_values_on_bars(ax)
plt.show()

# Plot accuracy by model and dataset with values on the bars
plt.figure(figsize=(10, 6))
ax = sns.barplot(data=df, x='Dataset', y='Accuracy', hue='Model')
plt.title('Accuracy by Model and Dataset')
add_values_on_bars(ax)
plt.show()

# Group tasks into focused and open-ended
df['TaskType'] = df['Task'].apply(lambda x: 'Focused' if x in ['p1', 'p2'] else 'OpenEnded')

# Calculate the best model per dataset and task type
best_model_per_task = df.groupby(['Dataset', 'TaskType', 'Model'])['Accuracy'].mean().reset_index()
best_model_per_task = best_model_per_task.loc[best_model_per_task.groupby(['Dataset', 'TaskType'])['Accuracy'].idxmax()]

# Print the best model for each dataset and task type
print("Best model per dataset and task type:")
print(best_model_per_task)
```
